### Model from scratch using Pytorch

```python
# inputs temp, rainfall, humidity
inputs = np.array([
                   [73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70]
], dtype = np.float32)
```

```python
# appales and oranges in tons
targets = np.array([
                    [56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119]
], dtype = np.float32)
```

1- init weights and baises 

```python
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
inputs, targets
```

```python
w = torch.randn(size=(2,3), requires_grad= True)
b = torch.randn(size = (2,), requires_grad=True)
w,b
```

2- creating the model

```python
y_hat = inputs @ w.T + b
y_hat
```

```python
def model(inputs, w,b):
  return inputs @ w.T +b
```

```python
preds = model(inputs,w,b)
preds
```

3- calculating the Loss

```python
#mse
diff = targets - preds
diff_sqr = diff**2
m_sqr = torch.sum(diff_sqr)/diff_sqr.numel()
m_sqr
```

```python
def mse ( targets,preds):
  diff = targets-preds
  return torch.sum(diff**2)/diff.numel()
```

```python
loss = mse(targets,preds)
loss
```

4- calculating grads of weights and baises aftet that reseting them to zero

```python
# we create gradiant of the loss function
'''loss.backward()
w.grad , b.grad'''
```

```python
'''
w.grad.zero_()
b.grad.zero_()
w.grad,b.grad
'''
```

```python
loss.backward()
with torch.no_grad():
  w-= w.grad * 1e-5
  b-= b.grad * 1e-5
  w.grad.zero_()
  b.grad.zero_()
```

```python
# new w and b
w,b
```

5- training the model for several epochs

```python
for _ in range(400):
  preds = model(inputs, w,b)
  loss = mse(targets,preds)
  loss.backward()
  with torch.no_grad():
    w-= w.grad * 1e-5
    b-= b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()
  print(loss)
```

```python
import torch
import numpy as np 
```

```python
inputs = np.array([[73, 67, 43], 
                   [91, 88, 64], 
                   [87, 134, 58], 
                   [102, 43, 37], 
                   [69, 96, 70], 
                   [74, 66, 43], 
                   [91, 87, 65], 
                   [88, 134, 59], 
                   [101, 44, 37], 
                   [68, 96, 71], 
                   [73, 66, 44], 
                   [92, 87, 64], 
                   [87, 135, 57], 
                   [103, 43, 36], 
                   [68, 97, 70]], np.float32)
```

```python
targets = np.array([[56, 70], 
                    [81, 101], 
                    [119, 133], 
                    [22, 37], 
                    [103, 119],
                    [57, 69], 
                    [80, 102], 
                    [118, 132], 
                    [21, 38], 
                    [104, 118], 
                    [57, 69], 
                    [82, 100], 
                    [118, 134], 
                    [20, 38], 
                    [102, 120]], np.float32)
```

```python
inputs = torch.tensor(inputs)
targets = torch.tensor(targets)
```

Tensor dataset and Data loader

```python
from torch.utils.data import TensorDataset
```

```python
ds = TensorDataset(inputs,targets)
```

```python
ds[1:5]
```

```python
from torch.utils.data import DataLoader
```

```python
ds_batched = DataLoader(ds,5,True)
```

buliding model and init w and b

```python
model = torch.nn.Linear(3,2 )
```

```python
model.weight, model.bias
```

```python
list(model.parameters())
```

Loss

```python
from torch.nn import functional as F
```

```python
loss = F.mse_loss(model(inputs),targets)
```

```python
loss
```

Optimizer

```python
from torch.optim import SGD
```

```python
opt = SGD(model.parameters(), lr= 1e-5)
```

train the model

```python
for epoch in range(500):
  for x,y in ds_batched:
    # 1 -predict
    preds = model(x)
    # 2-calculate the loss
    loss = F.mse_loss(preds, y)
    # 3- calculate the gradiant
    loss.backward()
    #4- otimize w,b
    opt.step()
    #5- set gradiant to zero
    opt.zero_grad()
  print(f"{epoch+1}/500 loss = {loss}")
```
